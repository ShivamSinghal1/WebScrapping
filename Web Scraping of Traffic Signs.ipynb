```python
import requests
import bs4
url = "https://en.wikipedia.org/wiki/Road_signs_in_India"
res = requests.get(url)
soup = bs4.BeautifulSoup(res.content)
allsign = soup.findAll('ul',{"class" : "gallery mw-gallery-traditional"})
#allsign[0] contains Regulatory Signs, allsign[1] contains Warning Signs, allsign[2] contains Informatory Signs

img = allsign[0].findAll('div',{"class" : "thumb"})
text = allsign[0].findAll('div',{"class" : "gallerytext"})

filename = "./WebScrapping/Traffic Sign Dataset/traffic_sign.csv";
path_for_img = "./WebScrapping/Traffic Sign Dataset/Image/sign{}.jpg";

ind = 0
with open(filename,'a') as file:
    #write column name
    file.write("id,title,type\n")
    #write row
    typ = "Regulatory Signs" 
    for t in text:
        title = t.p.text[:-1]
        row = "sign" + str(ind) + "," + title + "," + typ + "\n"
        file.write(row)
        ind+=1

ind = 0
for im in img:
    with open(path_for_img.format(ind),"wb") as file:
        imgUrl = "https:"
        imgUrl += im.img["src"]
        response= requests.get(imgUrl)
        file.write(response.content)
        ind+=1
```
